# Tool 3. AWS Translation

```python
# Import required libraries and set up our environment
from pprint import PrettyPrinter

import boto3

print("üìö Setting up the environment...")
pp = PrettyPrinter(indent=2)
translate = boto3.client("translate")
print("‚úÖ Environment setup complete!")
print(f"üåç Using AWS region: {translate.meta.region_name}")
```

```python
# Simple translation from Spanish to English
print("üîÑ Translating from Spanish to English...")

spanish_text = "MERECEMOS - VIVIR SIN MIEDO"
try:
    response = translate.translate_text(Text=spanish_text, SourceLanguageCode="es", TargetLanguageCode="en")

    print("\nüìù Translation details:")
    print(f"Original text (Spanish): {spanish_text}")
    print(f"Translated text (English): {response['TranslatedText']}")

    print("\nüì¶ Raw response from AWS:")
    pp.pprint(response)
except Exception as e:
    print(f"‚ùå Error during translation: {str(e)}")
```
